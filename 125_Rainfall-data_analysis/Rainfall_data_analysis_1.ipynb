### **Calculate the total, average, and variance of monthly rainfall.**

```python
months= ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
monthly_sum= rainfall_data.sum(axis=0)
monthly_sum=list(monthly_sum)[:12]
monthly_sum= pd.DataFrame({"monthly_sum": monthly_sum}, index=months)
monthly_sum
```

```python
montly_avg= rainfall_data.mean(axis=0)
montly_avg=list(montly_avg)[:12]
montly_avg=pd.DataFrame({"Avg_Rainfall": montly_avg}, index=months)
montly_avg
```

```python
montly_variance= rainfall_data.var(axis=0)
montly_variance=list(montly_variance)[:12]
montly_variance=pd.DataFrame({"Variance": montly_variance}, index=months)
montly_variance
```

```python
monthly_std= rainfall_data.std(axis=0)
monthly_std=list(monthly_std)[:12]
monthly_std = pd.DataFrame({"Months": months, "Std": monthly_std})
monthly_std
```

```python
rainfall_data["Yearly_Avg"]= rainfall_data.iloc[:, :12].mean(axis=1)
rainfall_data["Yearly_Variance"]= rainfall_data.iloc[:, :12].var(axis=1)
rainfall_data.head()
```

# **Trend Analysis**


### **• Visualize the long-term trend of annual rainfall over 125 years.** 
### **• Compute a 10-year rolling average of annual rainfall and overlay it on the trend plot.** 

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,8))
sns.lineplot(x=rainfall_data.index, y=rainfall_data["Total"], data=rainfall_data, palette="viridis")
```

```python
rainfall_data['10_Year_Rolling_Avg'] = rainfall_data["Total"].rolling(window=10, min_periods=1).mean()
```

```python
rainfall_data['10_Year_Rolling_Avg']
pd.DataFrame(rainfall_data['10_Year_Rolling_Avg'])
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x=rainfall_data.index, y=rainfall_data["Total"], label="Total Rainfall", color="blue")
sns.lineplot(x=rainfall_data.index, y=rainfall_data['10_Year_Rolling_Avg'], label="10-Year Rolling Avg", color="red")
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.title("Total Rainfall Trend and 10-Year Moving Average")
plt.legend()
plt.show()
```

# **Seasonal Analysis** 


### **Group the data by month and calculate the monthly average rainfall over 125 years, show graphically.** 

```python
months= rainfall_data[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']]
plt.figure(figsize=(12,8))
sns.barplot(x=months.columns, y=months.mean(), palette="viridis")
```

### **Identify the most and least variable months using the coefficient of variation (CV).**

```python
monthly_CV=[]
for i in range(12):
    cv= np.std(months.iloc[:, i])/np.mean(months.iloc[:, i])
    monthly_CV.append(cv)

print(monthly_CV)
```

```python
most_variable_month = rainfall_data.columns[monthly_CV.index(max(monthly_CV))]
least_variable_month = rainfall_data.columns[monthly_CV.index(min(monthly_CV))]
print({'most_variable_month': [most_variable_month, max(monthly_CV)], 'least_variable_month': [least_variable_month, min(monthly_CV)]})
```
